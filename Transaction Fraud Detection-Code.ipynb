4. Verify Data Quality

Why this step?

This step checks whether all transaction types belong to valid categories.

Observation:

All transaction records belong to valid transaction categories.

```python
valid_types = ['DEBIT', 'CASH_IN', 'CASH_OUT', 'TRANSFER']

invalid_types = df[~df['Type'].isin(valid_types)]

print("Invalid Transaction Types:")
print(invalid_types)

print("Count:", len(invalid_types))
```

Origin Balance Consistency

Why this step?

This step verifies whether original account balances are mathematically correct.

```python
balance_mismatch_org = df[
    (df['Oldbalance-Org'] - df['Amount']) != df['Newbalance-Org']
]

print(balance_mismatch_org.head())
print("Count:", len(balance_mismatch_org))
```

Destination Balance Consistency

Why this step?

This step verifies whether destination account balances are mathematically correct.

```python
balance_mismatch_dest = df[
    (df['Oldbalance-Dest'] + df['Amount']) != df['Newbalance-Dest']
]

print(balance_mismatch_dest.head())
print("Count:", len(balance_mismatch_dest))
```

Zero Amount Transactions

Why this step?

This step identifies transactions with zero transaction amount.

Observation:

No zero amount transactions were found.

```python
zero_amount = df[df['Amount'] == 0]

print(zero_amount)
print("Count:", len(zero_amount))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    roc_curve
)
```

```python
# Predicted class labels
y_pred = model.predict(X_test)

# Predicted probabilities
y_pred_proba = model.predict_proba(X_test)[:,1]
```

 2. Calculating Precision Score

here, we are checking the actual fraud transactions which are predicted as fraud in the tain-test data.

```python
precision = precision_score(y_test, y_pred)

print("Precision Score:", round(precision,4))
```

The model achieved a Precision Score of 1.00, indicating that every transaction predicted as fraudulent was actually fraudulent. This means the model generated zero false positives, minimizing unnecessary fraud alerts and reducing the risk of legitimate transactions being incorrectly flagged.

 3. Calculating Recall Score

```python
recall = recall_score(y_test, y_pred)

print("Recall Score:", round(recall,4))
```

The model achieved a Recall Score of 1.00, indicating that it successfully identified all fraudulent transactions in the test dataset. This means there were zero false negatives, ensuring that no fraud cases were missed during prediction.

 4. Calculating F1 score

here, f1 score balances the precision and recall scores

```python
f1 = f1_score(y_test, y_pred)

print("F1 Score:", round(f1,4))
```

The model achieved an F1-Score of 1.00, indicating a perfect balance between Precision and Recall. This means the model was able to identify all fraudulent transactions while avoiding any incorrect fraud predictions. The score reflects outstanding overall classification performance and demonstrates that the model is highly effective in detecting fraud without compromising prediction accuracy.

 5. Calculating ROC-AUC Score

here, this ROC-AUC score checks how well the model seperates fraud transactions to non-fraud transactions.

```python
roc_auc = roc_auc_score(y_test, y_pred_proba)

print("ROC-AUC Score:", round(roc_auc,4))
```

The model achieved a ROC-AUC Score of 1.00, indicating perfect discrimination between fraudulent and legitimate transactions. This means the model can completely separate the two classes and consistently assign higher fraud probabilities to fraudulent transactions than to legitimate ones. Such a score represents exceptional predictive capability and suggests that the model performs extremely well across different classification thresholds.
